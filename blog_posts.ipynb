```python
from transformers import pipeline

# Load models
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
sentiment_pipeline = pipeline("sentiment-analysis")

# Example blog post
blog_post = """
Artificial Intelligence is rapidly transforming industries. Many companies embrace AI
for efficiency and innovation, while others fear job losses and ethical concerns.
Governments are struggling to regulate it effectively. Overall, society seems divided
on whether AI will be more beneficial or harmful in the long run.
"""

# Step 1: Summarize the blog post
summary = summarizer(blog_post, max_length=60, min_length=20, do_sample=False)
summary_text = summary[0]['summary_text']

# Step 2: Sentiment analysis on the summary
sentiment = sentiment_pipeline(summary_text)

print("Blog summary:", summary_text)
print("Sentiment:", sentiment)
```
